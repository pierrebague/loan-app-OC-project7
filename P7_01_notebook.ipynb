```python
# BayesSearchCV
lightgbm_features_save = 'C:/Users/pierr/Documents/OC/projet7/script/features_saved_model2_02.csv'
lightgbm_features_save_model2_02 = pd.read_csv(lightgbm_features_save,low_memory=False)
print("fichier de",lightgbm_features_save_model2_02.shape[0],"produits et",lightgbm_features_save_model2_02.shape[1],"variables")
lightgbm_features_save_model2_02.head()
```

```python
del_features = ['TARGET', 'SK_ID_CURR', 'SK_ID_BUREAU', 'SK_ID_PREV', 'index', 'level_0']
predictors = list(filter(lambda v: v not in del_features, lightgbm_features_save_model2_02.columns))
```

```python
## EARLY_STOPPING a 100 
## https://github.com/scikit-optimize/scikit-optimize/issues/728
from skopt import BayesSearchCV
from skopt.space import Real, Integer
from lightgbm import LGBMClassifier
from sklearn.model_selection import StratifiedKFold

space = {
        'learning_rate': Real(0.001, 1.0, 'log-uniform'),
        'n_estimators': Integer(30, 5000),
        'num_leaves': Integer(2, 512),
        'max_depth': Integer(-1, 12),
        'min_child_samples': Integer(1, 256),
        'max_bin': Integer(100, 1000),
        'subsample': Real(0.01, 1.0, 'uniform'),
        'subsample_freq': Integer(0, 10),
        'colsample_bytree': Real(0.01, 1.0, 'uniform'),
        'min_child_weight': Real(0.01, 10.0, 'uniform'),
        'reg_lambda': Real(1e-9, 100.0, 'log-uniform'),
        'reg_alpha': Real(1e-9, 100.0, 'log-uniform'),
        'colsample_bytree': Real(0.01, 1.0, 'uniform'),
        'min_split_gain': Real(0.01, 1.0, 'uniform')           
        }

folds = StratifiedKFold(n_splits= 10, shuffle=True, random_state= 14)


## new_list = [function(item) for item in list if condition(item)]
for (train_idx, valid_idx) in (folds.split(df[predictors], df['TARGET'])):
    first_train_idx2,first_valied_idx2 = train_idx, valid_idx
    break
for (train_idx, valid_idx) in (folds.split(df[predictors].iloc[first_valied_idx2], df['TARGET'].iloc[first_valied_idx2])):
    first_train_idx,first_valied_idx = train_idx, valid_idx
    break
train_x, train_y = df[predictors].iloc[first_train_idx], df['TARGET'].iloc[first_train_idx]
valid_x, valid_y = df[predictors].iloc[first_valied_idx], df['TARGET'].iloc[first_valied_idx]
```

```python
len(first_train_idx2)
```

```python
len(train_x)
```

```python
opt = BayesSearchCV(
    LGBMClassifier(boosting_type='gbdt',subsample=0.1,is_unbalance=False),
    {
        **space
    },
    fit_params={
                'early_stopping_rounds': 100
                },
    verbose=2,
    cv=10,
    n_iter=100,
    n_jobs = 4
)

opt.fit(train_x, train_y)

print("val. score: %s" % opt.best_score_)
print("test score: %s" % opt.score(valid_x, valid_y))
```
